# Task 3: Dashboard â€“ Product Data Visualization
Visualize product categories and price distribution using bar charts and histograms.

```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# Step 1: Upload the CSV file
uploaded = files.upload()
df = pd.read_csv("products-100.csv")

# Step 2: Preview key columns
print(df[['Name', 'Category', 'Price', 'Stock']].head())
```

```python

# Step 3: Bar chart â€“ Product category distribution
df['Category'].value_counts().plot(kind='bar', color='skyblue', title='Product Categories')
plt.ylabel("Count")
plt.xlabel("Category")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

# Step 4: Histogram â€“ Product price distribution
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df['Price'].dropna().hist(bins=10, color='lightgreen')
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.show()
```
